```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Initialize the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")

def generate_response(input_text):
    # Encode the new user input, add the eos_token_id and return a tensor in Pytorch
    new_user_input_ids = tokenizer.encode(input_text + tokenizer.eos_token, return_tensors='pt')

    # Generate a chatbot response
    bot_input_ids = new_user_input_ids

    # The chatbot generates a response till it encounters the eos_token_id (end of sentence)
    chatbot_output = model.generate(bot_input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)

    # Decode the chatbot response
    output_text = tokenizer.decode(chatbot_output[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)

    return output_text

while True:
    user_input = input("User: ")
    print("Bot:", generate_response(user_input))
```

```python
import openai
import streamlit as st

# Set your API key
openai.api_key = 'your-api-key'

# Define the function to get the response from GPT-3
def get_response(input_text):
    response = openai.Completion.create(
      engine="text-davinci-002",
      prompt=input_text,
      temperature=0.5,
      max_tokens=100
    )
    return response.choices[0].text.strip()

# Set the title of your app
st.title('My Chatbot')

# Get the user input
input_text = st.text_input('You: ')

# Get the response and display it when the user presses the 'Talk' button
if st.button('Talk'):
    response = get_response(input_text)
    st.markdown(f'Chatbot: {response}')
```
